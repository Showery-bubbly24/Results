- **fixed acidity (ру. фиксированная кислотность)** - это характеристика, определяемая содержанием нелетучих кислот, таких как винная, яблочная и лимонная. Она важна для сбалансированности вкуса вина, придает ему свежесть и участвует в формировании его структуры. Оптимальный уровень фиксированной кислотности важен для создания гармоничного вина. Слишком высокая кислотность может сделать вино слишком резким, а слишком низкая – "безжизненным". 

- **volatile acidity (ру. летучая кислотность)** - это показатель содержания в вине кислот, которые могут испаряться с водяным паром. В основном это уксусная кислота и ее эфиры, которые придают вину неприятный запах и вкус, особенно в высоких концентрациях. В небольших количествах летучие кислоты могут придавать вину сложность и пикантность, но при избытке они становятся дефектом, ухудшая вкус и аромат. 

- **citric acid (ру. лимонная кислота)** - используется для корректировки кислотности, улучшения вкуса и предотвращения помутнений, связанных с железом. Хотя лимонная кислота присутствует в винограде в небольших количествах, её часто добавляют в вино, особенно на завершающих этапах производства, для достижения желаемой кислотности и вкусового баланса. Однако, стоит учитывать, что лимонная кислота может подвергаться молочнокислому брожению, в результате которого образуется уксусная кислота, придающая вину неприятный уксусный запах.

- **residual sugar (ру. остаточный сахар)** - это количество сахара, которое остается после завершения процесса брожения, когда дрожжи перерабатывают сахар в спирт. Этот остаточный сахар определяет сладость вина и может варьироваться в зависимости от типа вина и процесса его производства. Чем больше остаточного сахара, тем слаще вино. 

- **chlorides (ру. хлориды)** - присутствуют в виде хлоридов кальция и натрия. Эти соединения, хотя и являются естественной частью химического состава вина, в повышенных концентрациях могут влиять на его вкус, стабильность и даже вызывать коррозию оборудования. Хлориды, в частности хлорид кальция, могут придавать вину более мягкий, округлый и сладковатый вкус. 

- **free sulfur dioxide (ру. свободный диоксид серы)** - это консервант, который используется для защиты вина от окисления и развития бактерий. Он также помогает сохранить свежесть, фруктовые ароматы и вкусовые качества вина. При правильном использовании диоксид серы не должен изменять вкус и аромат вина. Однако, избыток диоксида серы может придать вину неприятный запах и ухудшить его вкус.

- **total sulfur dioxide (ру. общий диоксид серы)** - это суммарное количество диоксида серы (SO2), как добавленного в процессе производства, так и образовавшегося в результате естественных процессов брожения. Он является консервантом, предотвращающим окисление и порчу вина, а также защищающим его от бактерий и дрожжей. 

- **density (ру. плотность)** - это характеристика, описывающая его "тело" или текстуру, то есть насколько густым и насыщенным оно ощущается на вкус. Плотность вина зависит от содержания в нем спирта, сахара и глицерина, а также других экстрактивных веществ. Плотные вина кажутся более насыщенными, полнотелыми и богатыми на вкус, в то время как более легкие вина кажутся более освежающими и менее концентрированными. 

- **pH (ру. не переводится)** - это показатель кислотности вина, который определяет концентрацию ионов водорода. Он измеряется по шкале от 0 до 14, где 7 - нейтральный, ниже 7 - кислый, выше 7 - щелочной. В вине pH обычно находится в диапазоне от 3 до 4, при этом белые вина обычно более кислые, чем красные. Более низкие значения pH повышают стабильность вина, делая его менее подверженным бактериальным проблемам и окислению. 

- **sulphates (ру. сульфаты)** - это соединения серы, которые используются в виноделии в качестве консервантов и антиоксидантов. Они предотвращают окисление вина и подавляют рост нежелательных микроорганизмов, таких как бактерии и дикие дрожжи. Все вина содержат сульфиты, как естественный побочный продукт ферментации, так и добавленные в процессе производства. 

- **alcohol (ру. спирт)** - этанол, образуется в результате брожения виноградного сока. Его количество (крепость) влияет на вкус, структуру и общее восприятие вина, а также на его физиологическое воздействие. Слишком много или слишком мало алкоголя может нарушить баланс вина, делая его несбалансированным и неприятным. 

- **quality (ру. качество)** - таргет переменная, описывающая качество вина  

К сожалению, мы не являемся знатоками в вине (даже никогда пробовали), но из всего выше перечисленного можно выявить некоторые мысли и гипотезы, которые мы вспослдствии проверим.

Это все гипотезы и мы их проверим чуть позже.
1) фиксированная кислотность - слишком высокие показатели данной характеристики должны снижать качество вина, но, при этом, и слишком низкие показатели данной характеристики тоже должны снижать качество вина (нужен баланс)
2) летучая кислотность - чем ниже показатель, тем лучше вино
3) лимонная кислота - в нашем случае у нас нет информации о том, подверглась ли лимонная кислота молочнокислому брожению или нет, поэтому это учитывать не будем, но в идеале, конечно, и это стоит учитывать. Однако, прямого влияния на качетсво вина лимонная кислота не показывает, так что смеем предположить, что тут нужно просто соблюдать баланс
4) остаточный сахар - прямого влияния остаточных сахар тоже не проявляет на уровень качества вина
5) хлориды - высокое сожержание отрицательно влияет на качество вина
6) свободный диоксид серы - высокое содержание отриуательно влияет на качество вина
7) общий диоксид серы - должен быть в оптимальном диапазоне: недостаток ведет к окислению, избыток - к неприятному запаху
8) плотность - прямого влияния на качество вина не замечено (на вкус да, но на качество нет)
9) pH - чем ниже показатель, тем лучше
10) сульфаты - в умеренных количествах улучшают стабильность вина, но избыток может придать неприятный химический привкус (нужен баланс)
11) спирт - слишком высокие показатели данной характеристики должны снижать качество вина, но, при этом, и слишком низкие показатели данной характеристики тоже должны снижать качество вина (нужен баланс)

```python
# Теперь нужно создать два типа наборов данных
columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 
           'chlorides', 'free sulfur dioxide', 'total sulfur dioxide',	'density',	
           'pH', 'sulphates', 'alcohol']


# 1) В данном наборе мы проведем winsorization
import numpy as np
from scipy.stats.mstats import winsorize


final_wins_wines = encoded_all_wines.copy()
for column in columns:
    final_wins_wines[column] = winsorize(final_wins_wines[column], limits=(0.1, 0.1))


# 2) В этом наборе данных мы не будем работать с выбросами и оставим все как есть
final_wthoutWins_wines = encoded_all_wines.copy()
```

# Создание первых нейросетей

Напишем базовую нейросеть для каждого набора данных и подберем лучшие параметры с помощью Optuna

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


# Нейросеть для решения регрессионных задач
class WineRegressor(nn.Module):
    def __init__(self, input_size, hidden_sizes, dropout=0.3):
        super().__init__()
        layers = []
        sizes = [input_size] + hidden_sizes
        for i in range(len(sizes) - 1):
            layers.append(nn.Linear(sizes[i], sizes[i+1]))
            layers.append(nn.BatchNorm1d(sizes[i+1]))
            layers.append(nn.LeakyReLU(0.01))
            layers.append(nn.Dropout(dropout))
        
        layers.append(nn.Linear(sizes[-1], 1))
        self.net = nn.Sequential(*layers)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, x):
        # returns a single continuous prediction per sample
        return self.net(x).squeeze(-1)
```

```python
from sklearn.metrics import mean_absolute_error

# Оценка для регрессии
def evaluate_regression(model, val_loader):
    model.eval()
    predictions = []
    targets = []

    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            predictions.extend(outputs.cpu().numpy())
            targets.extend(labels.cpu().numpy())

    return mean_absolute_error(targets, predictions)
```

## Проверяем модель на винсоризированных данных

```python
import matplotlib.pyplot as plt

def plot_pred_vs_true(y_true, y_pred, label=None):
    plt.figure(figsize=(6, 6))
    plt.scatter(y_true, y_pred, alpha=0.5)
    # Диагональ идеальных предсказаний
    mn, mx = min(y_true), max(y_true)
    plt.plot([mn, mx], [mn, mx], linestyle='--')
    plt.xlabel("Истинные значения")
    plt.ylabel("Предсказанные значения")
    if label:
        plt.title(f"Model: {label}")
    else:
        plt.title("Предсказанные vs Истинные")
    plt.grid(True)
    plt.show()
```

```python
import torch
import torch.nn as nn


# Базовая структура модели (от нее начнем)
class OptimizedModel(nn.Module):
    def __init__(self, input_size, hidden_sizes, output_size, dropout=0.3):
        super().__init__()
        layers = []
        sizes = [input_size] + hidden_sizes
        
        for i in range(len(sizes)-1):
            layers.extend([
                nn.Linear(sizes[i], sizes[i+1]),
                nn.BatchNorm1d(sizes[i+1]),
                nn.ReLU(),
                nn.Dropout(dropout)
            ])
        
        layers.append(nn.Linear(sizes[-1], output_size))
        self.net = nn.Sequential(*layers)

    
    def forward(self, x):
        return self.net(x)


# Функции для обучения и оценки
def train_model(model, train_loader, optimizer, epochs=50):
    criterion = nn.CrossEntropyLoss()
    model.train()
    
    for epoch in range(epochs):
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()



def evaluate_model(model, val_loader):
    model.eval()
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    return correct / total
```

## Новая структура нейросети

```python
class ResidualBlock(nn.Module):
    def __init__(self, size, dropout=0.3):
        super().__init__()
        self.linear1 = nn.Linear(size, size)
        self.norm1 = nn.LayerNorm(size)
        self.act1 = nn.LeakyReLU(negative_slope=0.01)
        self.dropout = nn.Dropout(dropout)
        self.linear2 = nn.Linear(size, size)
        self.norm2 = nn.LayerNorm(size)
        self.act2 = nn.LeakyReLU(negative_slope=0.01)

    def forward(self, x):
        residual = x
        out = self.linear1(x)
        out = self.norm1(out)
        out = self.act1(out)
        out = self.dropout(out)
        out = self.linear2(out)
        out = self.norm2(out)
        out += residual  # skip connection
        out = self.act2(out)
        return out

class ImprovedClassifier(nn.Module):
    def __init__(self, input_size, hidden_sizes, output_size, dropout=0.3):
        super().__init__()
        layers = []
        sizes = [input_size] + hidden_sizes
        for i in range(len(sizes) - 1):
            layers.append(nn.Linear(sizes[i], sizes[i+1]))
            layers.append(nn.BatchNorm1d(sizes[i+1]))
            layers.append(nn.LeakyReLU(0.01))
            layers.append(nn.Dropout(dropout))
            # insert residual block for deeper networks
            if i > 0:
                layers.append(ResidualBlock(sizes[i+1], dropout=dropout))
        layers.append(nn.Linear(sizes[-1], output_size))
        self.net = nn.Sequential(*layers)
        self._initialize_weights()

    def _initialize_weights(self):
        # Xavier initialization for all linear layers
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def forward(self, x):
        return self.net(x)
```

```python
import optuna


def set_seed(seed=42):
    import random, numpy as np
    torch.manual_seed(seed)
    random.seed(seed)
    np.random.seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)


# Основная функция оптимизации
def objective(trial):
    # Предлагаемые параметры
    n_layers = trial.suggest_int('n_layers', 1, 4)
    hidden_units = []
    
    for i in range(n_layers):
        hidden_units.append(trial.suggest_categorical(f'n_units_{i}', [64, 128, 256, 512]))
    
    # Создание модели с конкретными параметрами
    model = ImprovedClassifier(
        input_size=input_size,
        hidden_sizes=hidden_units,
        output_size=output_size,
        dropout=trial.suggest_float('dropout', 0.1, 0.5)
    )
    
    optimizer = torch.optim.Adam(
        model.parameters(),
        lr=trial.suggest_float('lr', 1e-5, 1e-3, log=True),
        weight_decay=trial.suggest_float('weight_decay', 1e-6, 1e-3, log=True)
    )
    
    train_model(model, w_train_dataloader, optimizer, epochs=50)
    val_acc = evaluate_model(model, w_test_dataloader)
    return val_acc

# Перед запуском Optuna нужно определить:
# 1. Размеры входных и выходных данных
input_size = 13
output_size = 10
```

- Такие фичи, как fixed acidity, citric acid, residual sugar, free sulfur dioxide, total sulfur dioxide, pH и sulphates имеют очень низкую кореляцию (можно попробовать просто удалить их и посмотреть, что будет с результатами модели)
- Можно попробовать обьединить или преобразовать атрибуты в новые фичи, после чего так же можно посмотреть на их кореляцию с таргет-переменной (и конечно же протестировать модель)

```python
# Создадим датасет в котором не будет низко-корелирующих фич
without_lowCorr_wines = final_wthoutWins_wines.drop({'fixed acidity', 'citric acid', 'residual sugar', 
                                                    'free sulfur dioxide', 'total sulfur dioxide', 'pH', 'sulphates'}, axis=1)


# Очередная подготовка данных
X_train, X_test, y_train, y_test = split_data(without_lowCorr_wines)

without_lowCorr_train_dl, without_lowCorr_test_dl = prepare_data(X_train, X_test, y_train, y_test)
```

```python
# И создадим еще один датасет, в который уже дополним новыми фичами и посмотрим на их кореляцию
withnewfeatures_wines = final_wthoutWins_wines.copy()

# Новая фича - Общая кислотность
withnewfeatures_wines['total_acidity'] = withnewfeatures_wines['fixed acidity'] + withnewfeatures_wines['volatile acidity'] + withnewfeatures_wines['citric acid']

# Новая фича - Бинарная фича (высокое содержание алкоголя?)
withnewfeatures_wines['high_alcohol'] = (withnewfeatures_wines['alcohol'] > withnewfeatures_wines['alcohol'].median()).astype(int)

# Новые фичи - Логарифмы низко-корелирующих признаков
withnewfeatures_wines['log_fixed acidity'] =  np.log(withnewfeatures_wines['fixed acidity'] + 1e-6)
withnewfeatures_wines['log_citric acid'] = np.log(withnewfeatures_wines['citric acid'] + 1e-6)
withnewfeatures_wines['log_residual sugar'] = np.log(withnewfeatures_wines['residual sugar'] + 1e-6)
withnewfeatures_wines['log_free sulfur dioxide'] = np.log(withnewfeatures_wines['free sulfur dioxide'] + 1e-6)
withnewfeatures_wines['log_total sulfur dioxide'] = np.log(withnewfeatures_wines['total sulfur dioxide'] + 1e-6)
withnewfeatures_wines['log_pH'] = np.log(withnewfeatures_wines['pH'] + 1e-6)
withnewfeatures_wines['log_sulphates'] = np.log(withnewfeatures_wines['sulphates'] + 1e-6)


# Новые фичи - Квадраты низко-корелирующих признаков
withnewfeatures_wines['sq_fixed acidity'] = withnewfeatures_wines['fixed acidity'] ** 2
withnewfeatures_wines['sq_citric acid'] = withnewfeatures_wines['citric acid'] ** 2
withnewfeatures_wines['sq_residual sugar'] = withnewfeatures_wines['residual sugar'] ** 2
withnewfeatures_wines['sq_free sulfur dioxide'] = withnewfeatures_wines['free sulfur dioxide'] ** 2
withnewfeatures_wines['sq_total sulfur dioxide'] = withnewfeatures_wines['total sulfur dioxide'] ** 2
withnewfeatures_wines['sq_pH'] = withnewfeatures_wines['pH'] ** 2
withnewfeatures_wines['sq_sulphates'] = withnewfeatures_wines['sulphates'] ** 2
```

```python
def objective(trial):
    n_layers = trial.suggest_int('n_layers', 1, 4)
    hidden_units = []
    for i in range(n_layers):
        hidden_units.append(trial.suggest_categorical(f'n_units_{i}', [64, 128, 256, 512]))
    
    model = WineRegressor(
        input_size=input_size,
        hidden_sizes=hidden_units,
        dropout=trial.suggest_float('dropout', 0.1, 0.5)
    )

    optimizer = torch.optim.Adam(
        model.parameters(),
        lr=trial.suggest_float('lr', 1e-5, 1e-3, log=True),
        weight_decay=trial.suggest_float('weight_decay', 1e-6, 1e-3, log=True)
    )

    criterion = torch.nn.L1Loss()  # MAE
    model.train()

    for epoch in range(50):
        for inputs, labels in without_lowCorr_train_dl:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    val_mae = evaluate_regression(model, without_lowCorr_test_dl)
    return val_mae  # для MAE минимизируем
```
